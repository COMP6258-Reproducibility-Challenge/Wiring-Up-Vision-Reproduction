```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import numpy as np

def brain_predictivity_score(scores, upper):
    mean = np.mean(scores)
    return (mean/upper)*100

# Example array of x and y values
full_model_scores = [0.27871833, 0.25100567, 0.48113654, 0.35322569, 0.07908685]
full_model_mean = np.mean(full_model_scores)

CT_V2 = [0.22912021, 0.17094814, 0.46537991, 0.35746001, 0.0735246]
CT_V4 = [0.26761871, 0.18121143, 0.46642035, 0.33721024, 0.07690507]
CT_IT = [0.2147975, 0.17117837, 0.43023303, 0.33620002, 0.0352282]
CT_V2_Score = brain_predictivity_score(CT_V2, full_model_mean)
CT_V4_Score = brain_predictivity_score(CT_V4, full_model_mean)
CT_IT_Score = brain_predictivity_score(CT_IT, full_model_mean)

DS_V2 = [0.24794546, 0.1557341, 0.44403023, 0.33712959, 0.06040784]
DS_V4 = [0.26332701, 0.15864671, 0.39874365, 0.2538718, 0.05711225]
DS_IT = [0.25001659, 0.15740685, 0.36449156, 0.1770875, 0.04936319]
DS_V2_Score = brain_predictivity_score(DS_V2, full_model_mean)
DS_V4_Score = brain_predictivity_score(DS_V4, full_model_mean)
DS_IT_Score = brain_predictivity_score(DS_IT, full_model_mean)

ct_x_values = [50.915816, 41.099240, 1.889256]
ct_y_values = [CT_V2_Score, CT_V4_Score, CT_IT_Score]

ds_x_values = [50.850280, 40.771560, 0.513000]
ds_y_values = [DS_V2_Score, DS_V4_Score, DS_IT_Score]

fig, ax = plt.subplots(dpi=200)

ax.grid(True, color='gainsboro', zorder=1)

# Create the plot
ax.plot(ct_x_values, ct_y_values, '--', color='c',  label='Critical Training', zorder=2)
ax.scatter(ct_x_values, ct_y_values, color='c', zorder=2)
ax.text(5, 90, "Critical Training", color='c', ha='left', va='center', zorder=2)

ax.plot(ds_x_values, ds_y_values, '--', color='black', label="Downstream Training", zorder=2)
ax.scatter(ds_x_values, ds_y_values, color='black', zorder=2)
ax.text(5, 67, "Downstream Training", color='black', ha='left', va='center', zorder=2)

ax.scatter(53.416616, 100, color='black', zorder=2)
ax.text(53.416616, 100, "Standard Training  ", ha='right', va='center')

ax.set_ylim(0, 105)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_color('lightgrey')
ax.spines['bottom'].set_color('lightgrey')
ax.spines['left'].set_linewidth(2)
ax.spines['bottom'].set_linewidth(2)
ax.yaxis.set_major_formatter(mtick.PercentFormatter())

# Set the x-axis scale to logarithmic
#plt.xscale('log')

# Add labels and title
ax.set_xlabel('Number of Trained Parameters [Million]')
ax.set_ylabel('Brain Predictivity [% of Starndard Training]')
ax.tick_params(axis='both', which='both', bottom=False, left=False)

fig.set_size_inches(5, 5)

# Display the plot
#plt.legend()
plt.show()
```
